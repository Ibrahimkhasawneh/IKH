```python
import pandas as pd
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_absolute_error
from sklearn.model_selection import train_test_split

the_data = '/content/diabetes.csv'
the_data=pd.read_csv(the_data)
the_data.head()
```

```python
from matplotlib import pyplot as plt
the_data['Pregnancies'].plot(kind='hist', bins=20, title='Pregnancies')
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python
from sklearn.model_selection import train_test_split
y = the_data['Outcome']
infection_features = ['Pregnancies','Glucose','BloodPressure','SkinThickness','Insulin','BMI','DiabetesPedigreeFunction','Age']
X = the_data[infection_features]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=2)


the_model = DecisionTreeRegressor(random_state=1)
the_model.fit(X_train, y_train)


print("Making predictions for the following 5 people:")
print(X_test.head())
print("The predictions are")
predictions = the_model.predict(X_test.head())
print(predictions)

y_test_pred = the_model.predict(X_test)
print("Mean Absolute Error:", mean_absolute_error(y_test, y_test_pred))
```

```python
importances = the_model.feature_importances_
feature_names = infection_features
feature_importance = pd.Series(importances, index=feature_names).sort_values(ascending=False)
print(feature_importance)

feature_importance.plot(kind='bar', title='Feature Importance')
plt.show()
```

```python
import pandas as pd
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt

the_data = '/content/diabetes.csv'
the_data = pd.read_csv(the_data)
the_data.head()


y = the_data['Outcome']
infection_features = ['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI', 'DiabetesPedigreeFunction', 'Age']
X = the_data[infection_features]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=2)

rf_model = RandomForestRegressor(random_state=1)
rf_model.fit(X_train, y_train)

print("Making predictions for the following 5 people:")
print(X_test.head())
print("The predictions are")
predictions = rf_model.predict(X_test.head())
print(predictions)
```
